```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "../Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
#Count uniques in SN
player_count = purchase_data["SN"].nunique()

#Save the above total as a data frame
player_count_df = pd.DataFrame({"Player Count":[player_count]})

player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Basic calculations, unique items, average price, total purchases, total revenue

unique_items = purchase_data["Item ID"].nunique()

avg_price = purchase_data["Price"].mean()

total_purchases = purchase_data["Purchase ID"].count()

total_revenue = purchase_data["Price"].sum()

#Create a Summary Table

purchase_summary_df = pd.DataFrame({"Unique Item Count":[unique_items],
                                   "Average Price":[avg_price],
                                   "Total Purchases":[total_purchases],
                                   "Total Revenue":[total_revenue]})
#Format appropriately
purchase_summary_df[["Average Price","Total Revenue"]] = purchase_summary_df[["Average Price","Total Revenue"]].applymap("${:,.2f}".format)

#Print Summary Table
purchase_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#First, total unique players, and count of gender thereafter. Drop duplicates to ensure accuracy. 

unique_players_df = purchase_data.drop_duplicates(subset = "SN")

gender_totals_df = unique_players_df["Gender"].value_counts()

#Calcluate percentage using above
gender_percent = gender_totals_df / unique_players_df["Gender"].count() * 100

#Summarize in Table, Format, Display

gender_summary_df = pd.DataFrame({"Total":gender_totals_df, "Percentage of Players":gender_percent})

gender_summary_df["Percentage of Players"] = gender_summary_df["Percentage of Players"] \
                                            .map("{0:.2f}%".format)

gender_summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Count: Number of purchases by gender
gender_purchase_count = purchase_data.groupby("Gender")["Item Name"].count()

#Calculate: Average price of purchases by gender
gender_avg_price = purchase_data.groupby("Gender")["Price"].mean()

#Calculate: Total revenue by gender
gender_total_rev = purchase_data.groupby("Gender")["Price"].sum()

#Calculate: Average amount spend by each person
gender_total_average = total_revenue / gender_totals_df
```

```python
#Summarize
purchase_summary_df = pd.DataFrame({"Purchase Count": gender_purchase_count,
                       "Average Purchase Price": gender_avg_price,
                       "Total Purchase Revenue": gender_total_rev,
                       "Average Total Per Person": gender_total_average})

#Currency Formatting
purchase_summary_df["Average Purchase Price"] = purchase_summary_df["Average Purchase Price"].apply("${:.2f}".format)
purchase_summary_df["Total Purchase Revenue"] = purchase_summary_df["Total Purchase Revenue"].apply("${:.2f}".format)
purchase_summary_df["Average Total Per Person"] = purchase_summary_df["Average Total Per Person"].apply("${:.2f}".format)

purchase_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Define Bins for age
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize using Age Bins
unique_players_df["Age Ranges"] = pd.cut(unique_players_df["Age"], age_bins, labels = age_groups)

# Basic Calculations
age_demo_totals = unique_players_df["Age Ranges"].value_counts()
age_demo_percents = age_demo_totals / player_count * 100

# Create the Data Frame
age_demo = pd.DataFrame({"Total Purchase Count": age_demo_totals, "Percentage of Players": age_demo_percents})

# Format everything
age_demo = age_demo.round(2)

# Display the DataFrame
age_demo.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Place purchase data into age bings
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels = age_groups)

# Basic Calculations
age_counts = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")
age_total_purchase = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Revenue")
age_avg = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
normalized = age_total_purchase / age_demo["Total Purchase Count"]


# Create new DataFrame
age_data = pd.DataFrame({"Purchase Count": age_counts, "Average Purchase Price": age_avg, "Total Purchase Revenue": age_total_purchase, "Normalized Totals": normalized})

# Reformatting the columns
age_data ["Purchase Count"] = age_data["Purchase Count"].map("{:,}".format)
age_data["Total Purchase Revenue"] = age_data["Total Purchase Revenue"].map("${:,.2f}".format)
age_data["Average Purchase Price"] = age_data["Average Purchase Price"].map("${:,.2f}".format)
age_data["Average Total Per Person"] = age_data["Normalized Totals"].map("${:,.2f}".format)
age_data = age_data.loc[:, ["Purchase Count", "Total Purchase Revenue", "Average Purchase Price", "Average Total Per Person"]]

# Display the new DataFrame
age_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Basic Calculations
user_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")
user_average = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_total = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Revenue")

# Create df
user_data = pd.DataFrame({"Purchase Count": user_count, "Average Purchase Price": user_average, "Total Purchase Revenue": user_total})

# Sort total purchase value (descending order)
user_sorted = user_data.sort_values("Total Purchase Revenue", ascending = False)

# Reformatting the columns
user_sorted["Average Purchase Price"] = user_sorted["Average Purchase Price"].map("${:,.2f}".format)
user_sorted["Total Purchase Revenue"] = user_sorted["Total Purchase Revenue"].map("${:,.2f}".format)
user_sorted = user_sorted.loc[:,["Purchase Count", "Average Purchase Price", "Total Purchase Revenue"]]

# Display the DataFrame
user_sorted.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Columns to Retrieve: Item Name, Item ID, and Item Price
item_data = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

# Calculations and groupby
purchase_count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
item_price = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"]
total_purchase_value = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Revenue")

# Create df
item_data_pd = pd.DataFrame({"Purchase Count": purchase_count, "Item Price": item_price, "Total Purchase Revenue": total_purchase_value})

# Sort purchase count (descending order)
item_data_count_sorted = item_data_pd.sort_values("Purchase Count", ascending = False)

# Format
item_data_count_sorted["Purchase Count"] = item_data_count_sorted["Purchase Count"].map("{:,}".format)
item_data_count_sorted["Item Price"] = item_data_count_sorted["Item Price"].map("${:,.2f}".format)
item_data_count_sorted["Total Purchase Revenue"] = item_data_count_sorted["Total Purchase Revenue"].map("${:,.2f}".format)
item_popularity = item_data_count_sorted.loc[:,["Purchase Count", "Item Price", "Total Purchase Revenue"]]

# Display df
item_popularity.head(10)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sort previous DataFrame by Total Purchase Rev
item_total_value = item_data_pd.sort_values("Total Purchase Revenue", ascending = False)

# Reformat the columns, again
item_total_value["Purchase Count"] = item_total_value["Purchase Count"].map("{:,}".format)
item_total_value["Item Price"] = item_total_value["Item Price"].map("${:,.2f}".format)
item_total_value["Total Purchase Revenue"] = item_total_value["Total Purchase Revenue"].map("${:,.2f}".format)
most_profitable = item_total_value.loc[:,["Purchase Count", "Item Price", "Total Purchase Revenue"]]

# Display the new DataFrame below
most_profitable.head(10)
```
